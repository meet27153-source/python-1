Polymorphism- overloading and overriding

.Allows programmers to write methods that can process a variety of different types of functionalities with the same name 
.overloading implies using numerous methods with the same name but accepting a different no. of arguments
.overriding implies child class overrides the parent class method of the same name,parameters and return type,it is used to overwrite a parent class method in derived class

```python
def sum_numbers(*args):
    result=0
    for i in args:
        result+=i
    print("sum:",result)
print("Similar to method overloading concept")
print("Single argument:",end=" ")
sum_numbers(10)
print("Double argument:",end=" ")
sum_numbers(10,2)
print("Multiple argument:",end=" ")
sum_numbers(1,2,3,4,5)
```

```python
def fun(a):
    print("x")
def fun(a,b):
    print("x,y")
fun(5)
```

```python
class A:
    def first(self):
        print("First function of class A")
    def second(self):
        print("Second function of class A")
class B(A):
    def first(self):
        print("Redfined function of class A in class B")
    def display(self):
        print("Display function of child class")
obj=B()
obj.first()
obj.second()
A().first()
```

```python
class A:
    def __init__(self,a):
        self.a=a
    def __add__(self,other):
        return self.a+other.a
obj1=A(1)
obj2=A(2)
obj3=A("Hello")
obj4=A("Hi")
print(obj1+obj2)
print(obj3+obj4)

print(A.__add__(obj1,obj2))
print(A.__add__(obj3,obj4))

print(obj1.__add__(obj2))
print(obj3.__add__(obj4))
```

```python
class A:
    def __init__(self,a):
        self.a=a
    def __gt__(self,other):
        if self.a>other.a:
            return True
        else:
            return False
    def __eq__(self,other):
        if self.a==other.a:
            return True
        else:
            return False
    def __mul__(self,other):
        return (self.a**other.a)
    def __truediv__(self,other):
        return (self.a/other.a)
obj1=A(5)
obj2=A(5)
if obj1>obj2:
    print("ob1 greater than obj2")
elif obj1<obj2:
    print("ob2 greater than obj1")
if obj1==obj2:
    print("ob1 equal obj2")
else:
    print("ob1 not equal to obj2")
print(obj1*obj2)
print(obj1/obj2)
```

Inheritance
1.Single Inheritance

```python
class parent:
    def func1(self):
        print("Hello Parent")
class child(parent):
    def func2(self):
        print("Hello Child")
obj1=parent()
obj2=child()
obj2.func1()
obj2.func2()
```

2.Multiple Inheritance

```python
class parent1:
    def func1(self):
        print("Hello Parent1")
class parent2:
    def func2(self):
        print("Hello Parent2")
class parent3:
    def func2(self):
        print("Hello Parent3")
class child(parent1,parent3,parent2):
    def func3(self):
        print("Hello child")
obj1=parent()
obj2=child()
obj2.func1()
obj2.func2()
obj2.func3()
print(child.__mro__)
```

3.Multilevel Inheritance

```python
class grandparent:
    def func1(self):
        print("Hello Grandparent")
class parent(grandparent):
    def func2(self):
        print("Hello Parent")
class child(parent):
    def func3(self):
        print("Hello Child")
obj1=parent()
obj2=child()
obj2.func1()
obj2.func2()
obj2.func3()
```
