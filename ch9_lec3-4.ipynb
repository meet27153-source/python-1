4.Hierarchical Inheritance

```python
class parent:
    def func1(self):
        print("Hello World")
class child1(parent):
    def func2(self):
        print("Hello Child1")
class child2(parent):
    def func3(self):
        print("Hello Child2")
obj1=child1()
obj2=child2()
obj1.func1()
obj1.func2()
obj2.func1()
obj2.func3()
```

5.Hybrid Inheritance

```python
class parent1:
    def func1(self):
        print("Parent1")
class parent2:
    def func2(self):
        print("Parent2")
class child1(parent1):
    def func3(self):
        print("Child1")
class child2(child1,parent2):
    def func1(self):
        print("Child2")
obj1=child1()
obj2=child2()
obj1.func1()
obj2.func3()
```

1.super()

```python
class parent:
    def __init__(self):
        self.attr1=50
        self.attr2=60
class child(parent):
    def __init__(self):
        super().__init__()
        self.attr3=70
obj=child()
print(obj.attr1)

class parent1:
    def display(self):
        print("Hello Parent")
class child1(parent1):
    def display(self):
        super().display()
        print("Hello Child")
obj1=child1()
obj1.display()
```

2.issubclass()

```python
class parent:
    def func1(self):
        print("Hello Parent")
class child(parent):
    def func2(self):
        print("Hello Child")
print(issubclass(child,parent))
print(issubclass(parent,child))
```

3.isinstance()

```python
class parent:
    def func1(self):
        print("Hello Parent")
class child(parent):
    def func2(self):
        print("Hello Child")
A=child()
B=parent()
print(isinstance(A,child))
print(isinstance(A,parent))
print(isinstance(B,child))
print(isinstance(B,parent))
```

Method Resolution Order(mro)
-Denotes the way a program Language resolves a method or attribute
-MRO defines the order in which the base class are searched when executing a method
-First,the method or attribute is searched within a class and then it follows the order we specified while inheritance

```python
class A:
    def rk(self):
        print("Class A")
class B(A):
    def rk(self):
        print("Class B")
class C(A):
    def rk(self):
        print("Class C")
class D(B,C):
    pass
obj=D()
obj.rk()
print(D.__mro__)
print(D.mro())
```

```python
class p1:
    def foo(self):
        print("Foo from p1")
class p2:
    def foo(self):
        print("Foo from p2")
    def bar(self):
        print("Bar from p2")
class c1(p1,p2):
    pass
class c2(p2,p1):
    def bar(self):
        print("Bar from c2")
class GC(c1,c2):
    pass
obj=GC()
obj.foo()
obj.bar()
GC.__mro__
```

Abstract Class

```python
from abc import ABC,abstractmethod
class shape(ABC):
    def __init__(self,shape_name):
        self.shape_name=shape_name
    @abstractmethod
    def draw(self):
        pass
    
class circle(shape):
    def __init__(self):
        super().__init__("circle")
    def draw_circle(self):
        print("Drawing:",self.shape_name)
    def draw(self):
        pass
obj=circle()
obj.draw_circle()
```

```python
class Book:
    def __init__(self,name,n,authors,publisher,ISBN,year):
        self.name=name
        self.n=n
        self.authors=authors
        self.publisher=publisher
        self.ISBN=ISBN
        self.year=year
    def display(self):
        print("Name:",self.name)
        print("No. of authors:",self.n)
        print("Authors:",self.authors)
        print("Publisher:",self.publisher)
        print("ISBN:",self.ISBN)
        print("Year:",self.year)
class CourseBook(Book):
    def __init__(self,name,n,authors,publisher,ISBN,year,course):
        super().__init__(name,n,authors,publisher,ISBN,year)
        self.course=course
    def display(self):
        super().display()
        print("Course:",self.course)
obj=CourseBook("A","3","Meet","Mop","2200","2006","cseds")
obj.display()
```
