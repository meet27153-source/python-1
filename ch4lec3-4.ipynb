```python
s="Py$@*?th45729on"
sum=0
sp=0
up=0
low=0
d=0
alphabet=0
for i in s:
    if(i.isalpha()):
        alphabet=alphabet+1
        if(i.isupper()):
            up=up+1
        elif(i.islower()):
            low=low+1
    elif(i.isdigit()):
        d=d+1
        sum=sum+int(i)
    else:
        sp=sp+1
print("Uppercase:",up)
print("Lowercase:",low)
print("Alphabet:",alphabet)
print("Digits",d)
print("Special character:",sp)
print("Sum:",sum)
print("Average",sum/d)
```
