```python
import matplotlib.pyplot as plt
import numpy as np
time=np.arange(12)
income=np.array([5,9,6,6,10,7,6,4,4,5,6,4])
expanse=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
plt.plot(time,income,color='green')
plt.plot(time,expanse,color='red')
plt.fill_between(time,income,expanse,where=(income>expanse),color='green',alpha=0.5,label='positive',interpolate=True)
plt.fill_between(time,income,expanse,where=(income<=expanse),color='red',alpha=0.5,label='negative',interpolate=True)
plt.xlabel('Time')
plt.ylabel('Income/Expanse Comaprison')
plt.title('Area filled between 2 lines')
plt.legend()
plt.show()
```
