concatenate()

```python
import numpy as np
arr1=np.array([[1,2],[3,4]])
arr2=np.array([[5,6],[7,8]])
arr=np.concatenate((arr1,arr2),axis=1)
print(arr)
```

```python
import numpy as np
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
arr3=np.array([7,8,9])
arr=np.concatenate((arr1,arr2,arr3))
print(arr)
```

array_split()

```python
import numpy as np
arr1=np.array([1,2,3,4,5,6])
new=np.array_split(arr1,7)
print(new)
```

```python
import numpy as np
arr1=np.array([[1,2],[3,4],[5,6],[7,8]])
new=np.array_split(arr1,3,axis=1)
print(new)
```

where()

```python
import numpy as np
arr1=np.array([1,2,3,4,5,4,7,4,9])
x=np.where(arr1==4)
print(x)
```

```python
import numpy as np
arr1=np.array([[1,2],[3,4],[4,5],[7,4]])
x=np.where(arr1%2==0)
print(x)
```

```python
import numpy as np
arr1=np.array([[[1,2,3]],[[4,5,7]],[[1,4,3]]])
x=np.where(arr1==4)
print(x)
```

sort()

```python
import numpy as np
arr1=np.array([1,2,3,4,5,4,3,2])
print(np.sort(arr1,axis=0))
arr2=np.array(['apple','cherry','straberry','banana'])
print(np.sort(arr2))
arr3=np.array([True,False,True])
print(np.sort(arr3))
```

```python
import numpy as np
arr1=np.array([[3,4,2],[1,7,3],[4,2,5]])
print(np.sort(arr1,axis=None))
```

```python
import numpy as np
arr1=np.array([[[3,4,2],[1,7,3]],[[5,8,7],[4,2,5]]])
print(np.sort(arr1,axis=1))
```

```python
import numpy as np
att=7
print("You have 7 attempts")
x=np.random.randint(100)
for i in range(att):
    a=int(input(("Guess no. from 1 to 100:")))
    att-=1
    print("Attempt left:",att)
    if (att==0):
        print("Game over,you lose.The number was:",x)
    elif (a==x):
        print("You guessed the number")
    elif (a>100):
        print("Invalid number")
```
