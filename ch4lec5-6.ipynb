WAP to check validity of a password.Primary condition for password validation
1]minimum 8 characters 2]Alphabets must be between [a-z] 3]Atleast one alphabet should be uppercase[A-Z]
4]Atleast 1 number between[0-9] 5]ATleast 1 chracter from[_ or @ or $]

```python
password=input("Enter a valid password:")
lower=0
upper=0
digit=0
sp=0
if(len(password)>=8):
    for i in password:
        if(i.islower()):
            lower+=1
        if(i.isupper()):
            upper+=1
        if(i.isdigit()):
            digit+=1
        if(i=="@" or i=="_" or i=="$"):
            sp+=1
if lower>=1 and upper>=1 and sp>=1 and digit>=1 and (lower+upper+digit+sp)==len(password):
      print("Valid password")
else:
      print("Invalid password")
```
